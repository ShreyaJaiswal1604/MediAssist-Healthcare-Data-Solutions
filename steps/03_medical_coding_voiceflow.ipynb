```python
import re
import torch
import spacy
from transformers import AutoTokenizer, BertForSequenceClassification

```

```python
# Load the pretrained model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("AkshatSurolia/ICD-10-Code-Prediction")
model = BertForSequenceClassification.from_pretrained("AkshatSurolia/ICD-10-Code-Prediction")
config = model.config

```

```python

# Load SciSpaCy model for medical term normalization
nlp = spacy.load("en_core_sci_md")  # Requires SciSpaCy model installation
```

```python
def structure_clinical_notes(text):
    """
    Categorize clinical notes into structured sections.
    """
    print("Structuring clinical notes...")
    # Clean the text
    cleaned_text = text.replace("___", "").strip()

    # Define regex patterns for section headers
    sections = {
        "Patient Demographics": r"(Name:.*?Sex:.*?Service:.*?Allergies:.*?)Chief Complaint:",
        "Chief Complaint": r"Chief Complaint:(.*?)Major Surgical or Invasive Procedure:",
        "Major Procedures": r"Major Surgical or Invasive Procedure:(.*?)History of Present Illness:",
        "History of Present Illness": r"History of Present Illness:(.*?)Past Medical History:",
        "Discharge Diagnosis": r"Discharge Diagnosis:(.*?)Discharge Condition:"
    }

    structured_data = {}
    for section, pattern in sections.items():
        match = re.search(pattern, cleaned_text, re.S)
        structured_data[section] = match.group(1).strip() if match else ""

    print("Clinical notes structured successfully!")
    print(f"Structured Data: {structured_data}")
    return structured_data



```

```python
def refine_clinical_data(structured_data):
    """
    Refine extracted clinical data for accurate ICD mapping.
    """
    print("\nRefining structured data...")
    refined_data = {}
    for section, text in structured_data.items():
        print(f"\nProcessing section: {section}")
        
        # Normalize terms and clean text
        cleaned_text = text.replace("___", "").strip()
        
        if section in ["Chief Complaint", "Discharge Diagnosis"]:
            refined_data[section] = split_multi_entries(cleaned_text)  # Split into multiple diagnoses
            print(f"Refined {section}: {refined_data[section]}")
        
        elif section == "Major Procedures":
            # Clean procedures by removing colons and splitting
            refined_procedures = clean_colons_and_tokens(cleaned_text)
            refined_data[section] = refined_procedures
            print(f"Refined Major Procedures: {refined_data[section]}")
        
        elif section == "History of Present Illness":
            # Extract key clinical details
            refined_data[section] = extract_key_information(cleaned_text)
            print(f"Refined History of Present Illness: {refined_data[section]}")
        
        else:
            # General cleaning for other sections
            refined_data[section] = cleaned_text
            print(f"Refined {section}: {refined_data[section]}")
    
    return refined_data
```

```python
def normalize_terms(text):
    """
    Normalize medical terms using SciSpaCy or similar tools.
    """
    print(f"Normalizing terms for text: {text}")
    doc = nlp(text)
    normalized = [ent.text for ent in doc.ents if ent.label_ in ["DISEASE", "SYMPTOM"]]
    return ", ".join(normalized)
```

```python
def split_multi_entries(text):
    """
    Split multiple diagnoses or procedures into individual entries.
    """
    if not text:  # Handle None or empty input
        print("No text provided for splitting.")
        return []
    print(f"Splitting entries for text: {text}")
    return [entry.strip() for entry in re.split(r"[;\n]", text) if entry.strip()]
```

```python
def extract_key_information(text):
    """
    Extract key diagnoses or conditions from verbose descriptions.
    """
    print(f"Extracting key information from text: {text}")
    doc = nlp(text)
    keywords = [ent.text for ent in doc.ents if ent.label_ in ["DISEASE", "SYMPTOM"]]
    return ", ".join(keywords)
```

```python
def predict_icd(text, tokenizer, model, config):
    """
    Predict the ICD code for a given text using the loaded model.
    """
    if not text.strip():
        return "No text provided for prediction"
    print(f"Predicting ICD code for text: {text}")
    tokens = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)
    outputs = model(**tokens)
    predicted_code_idx = torch.argmax(outputs.logits, dim=1).item()
    icd_code = config.id2label[predicted_code_idx]
    print(f"Predicted ICD Code: {icd_code}")
    return icd_code
```

```python
def map_multiple_icd(entries, tokenizer, model, config):
    """
    Map a list of entries to their respective ICD codes.
    """
    icd_mappings = {}
    for entry in entries:
        icd_code = predict_icd(entry, tokenizer, model, config)
        icd_mappings[entry] = icd_code
        print(f"Entry: {entry} -> ICD Code: {icd_code}")
    return icd_mappings
```

```python
def clean_entries(entries):
    """
    Remove leading colons or unnecessary characters from entries.
    """
    print(f"Original Entries: {entries}")
    cleaned_entries = [entry.lstrip(":").strip() for entry in entries]
    print(f"Cleaned Entries: {cleaned_entries}")
    return cleaned_entries
```

```python

def process_and_map_icd(refined_notes, tokenizer, model, config):
    """
    Process refined notes and predict ICD codes for diagnoses and procedures.
    """
    print("\n--- Processing Diagnoses and Procedures ---")
    
    diagnoses_entries = refined_notes.get("Chief Complaint", []) + refined_notes.get("Discharge Diagnosis", [])
    procedures_entries = refined_notes.get("Major Procedures", [])
    
    print(f"\nDiagnoses Entries: {diagnoses_entries}")
    print(f"Procedures Entries: {procedures_entries}")
    
    print("\n--- Mapping Diagnoses to ICD Codes ---")
    diagnoses_icd_mapping = map_multiple_icd(diagnoses_entries, tokenizer, model, config)
    
    print("\n--- Mapping Procedures to ICD Codes ---")
    procedures_icd_mapping = map_multiple_icd(procedures_entries, tokenizer, model, config)
    
    print("\n--- Final Mappings ---")
    print(f"Diagnoses ICD Codes: {diagnoses_icd_mapping}")
    print(f"Procedures ICD Codes: {procedures_icd_mapping}")
    return diagnoses_icd_mapping, procedures_icd_mapping
```

```python
# Ensure clean_colons_and_tokens and split_multi_entries handle entries properly.
def clean_colons_and_tokens(text):
    """
    Remove leading colons and unnecessary characters/tokens from text.
    """
    print(f"Cleaning text: {text}")
    cleaned_entries = [entry.lstrip(":").strip() for entry in re.split(r"[;\n]", text) if entry.strip()]
    print(f"Cleaned entries: {cleaned_entries}")
    return cleaned_entries
```

```python
def split_multi_entries(text):
    """
    Split multiple diagnoses or procedures into individual entries.
    """
    print(f"Splitting entries for text: {text}")
    entries = [entry.strip() for entry in re.split(r"[;\n]", text) if entry.strip()]
    print(f"Original Entries: {entries}")
    return entries
```

```python
import re
import torch
import spacy
from transformers import AutoTokenizer, BertForSequenceClassification, AutoModel, AutoTokenizer

# Load SciSpaCy model for medical term normalization
nlp = spacy.load("en_core_sci_md")  # Requires SciSpaCy model installation

# Load the BERT-based model
bert_tokenizer = AutoTokenizer.from_pretrained("AkshatSurolia/ICD-10-Code-Prediction")
bert_model = BertForSequenceClassification.from_pretrained("AkshatSurolia/ICD-10-Code-Prediction")
bert_config = bert_model.config

# Load the Llama-based model
llama_model_name = "ContactDoctor/Bio-Medical-MultiModal-Llama-3-8B-V1"
llama_tokenizer = AutoTokenizer.from_pretrained(llama_model_name, trust_remote_code=True)
llama_model = AutoModel.from_pretrained(llama_model_name, trust_remote_code=True)

def structure_clinical_notes(text):
    # Define regex patterns for section headers
    sections = {
        "Patient Demographics": r"(Name:.*?Sex:.*?Service:.*?Allergies:.*?)Chief Complaint:",
        "Chief Complaint": r"Chief Complaint:(.*?)Major Surgical or Invasive Procedure:",
        "Major Procedures": r"Major Surgical or Invasive Procedure:(.*?)History of Present Illness:",
        "History of Present Illness": r"History of Present Illness:(.*?)Past Medical History:",
        "Discharge Diagnosis": r"Discharge Diagnosis:(.*?)Discharge Condition:"
    }

    structured_data = {}
    for section, pattern in sections.items():
        match = re.search(pattern, text, re.S)
        structured_data[section] = match.group(1).strip() if match else ""
    return structured_data

def refine_clinical_data(structured_data):
    refined_data = {}
    for section, text in structured_data.items():
        cleaned_text = text.replace("___", "").strip()
        if section in ["Chief Complaint", "Discharge Diagnosis"]:
            refined_data[section] = split_multi_entries(cleaned_text)
        elif section == "Major Procedures":
            refined_data[section] = clean_entries(split_multi_entries(cleaned_text))
        else:
            refined_data[section] = cleaned_text
    return refined_data

def predict_icd_bert(text, tokenizer, model, config):
    if not text.strip():
        return "No text provided for prediction"
    tokens = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)
    outputs = model(**tokens)
    predicted_code_idx = torch.argmax(outputs.logits, dim=1).item()
    return config.id2label[predicted_code_idx]

def predict_icd_llama(text, tokenizer, model):
    if not text.strip():
        return "No text provided for prediction"
    tokens = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)
    outputs = model(**tokens)
    logits = outputs.logits  # Assuming logits are returned
    predicted_code_idx = torch.argmax(logits, dim=1).item()
    return f"ICD-{predicted_code_idx}"  # Placeholder mapping

def map_multiple_icd(entries, tokenizer, model, config=None, model_type="bert"):
    icd_mappings = {}
    for entry in entries:
        if model_type == "bert":
            icd_code = predict_icd_bert(entry, tokenizer, model, config)
        elif model_type == "llama":
            icd_code = predict_icd_llama(entry, tokenizer, model)
        icd_mappings[entry] = icd_code
    return icd_mappings

def clean_entries(entries):
    return [entry.lstrip(":").strip() for entry in entries]

def split_multi_entries(text):
    return [entry.strip() for entry in re.split(r"[;\n]", text) if entry.strip()]

def process_and_map_icd(structured_notes, tokenizer, model, config=None, model_type="bert"):
    diagnoses_text = structured_notes.get("Chief Complaint", "") + " " + structured_notes.get("Discharge Diagnosis", "")
    procedures_text = structured_notes.get("Major Procedures", "")

    diagnoses_icd_mapping = map_multiple_icd(split_multi_entries(diagnoses_text), tokenizer, model, config, model_type)
    procedures_icd_mapping = map_multiple_icd(split_multi_entries(procedures_text), tokenizer, model, config, model_type)
    return diagnoses_icd_mapping, procedures_icd_mapping

# Clinical Notes
clinical_notes = """
Name:  ___               Unit No:   ___
Admission Date:  ___              Discharge Date:   ___
Date of Birth:  ___             Sex:   M
Service: MEDICINE
Allergies: Amoxicillin / azithromycin
Chief Complaint:
Liver Failure
Major Surgical or Invasive Procedure:
Bone marrow biopsy; Liver biopsy; EGD
Discharge Diagnosis:
Acute Liver Failure; Hemophagocytic Lymphohistiocytosis; HSV 1
"""

# Structure and Refine Clinical Notes
structured_notes = structure_clinical_notes(clinical_notes)
refined_notes = refine_clinical_data(structured_notes)

# Predict ICD codes using BERT
bert_diagnoses_icd, bert_procedures_icd = process_and_map_icd(refined_notes, bert_tokenizer, bert_model, bert_config, "bert")
print("\n--- BERT Model Results ---")
print(f"Diagnoses ICD Codes: {bert_diagnoses_icd}")
print(f"Procedures ICD Codes: {bert_procedures_icd}")

# Predict ICD codes using Llama
llama_diagnoses_icd, llama_procedures_icd = process_and_map_icd(refined_notes, llama_tokenizer, llama_model, model_type="llama")
print("\n--- Llama Model Results ---")
print(f"Diagnoses ICD Codes: {llama_diagnoses_icd}")
print(f"Procedures ICD Codes: {llama_procedures_icd}")

# Compare Results
print("\n--- Comparison of Results ---")
print("BERT Diagnoses vs. Llama Diagnoses:")
for key in bert_diagnoses_icd:
    print(f"{key}: BERT -> {bert_diagnoses_icd[key]}, Llama -> {llama_diagnoses_icd.get(key, 'N/A')}")
print("\nBERT Procedures vs. Llama Procedures:")
for key in bert_procedures_icd:
    print(f"{key}: BERT -> {bert_procedures_icd[key]}, Llama -> {llama_procedures_icd.get(key, 'N/A')}")
```

```python
import transformers
import torch

# Load the tokenizer and model
tokenizer = transformers.LlamaTokenizer.from_pretrained('axiong/PMC_LLaMA_13B')
model = transformers.LlamaForCausalLM.from_pretrained('axiong/PMC_LLaMA_13B')
model.cuda()  # Move the model to GPU

# Define the medical text input
medical_text = """
Patient History:
The patient presents with acute liver failure and hemophagocytic lymphohistiocytosis.
Procedures performed include bone marrow biopsy and liver biopsy.
Diagnosis:
Acute liver failure due to HSV-1 infection.
"""

# Format the input prompt to request ICD codes
input_prompt = (
    f"{medical_text}\n\n"
    "Please provide the ICD-10 medical codes for the above patient case."
)

# Tokenize the input text
inputs = tokenizer(input_prompt, return_tensors="pt").to("cuda")

# Generate text
output_tokens = model.generate(
    inputs['input_ids'],
    max_length=200,  # Adjust length as needed
    num_return_sequences=1,  # Generate one response
    temperature=0.7,  # Control randomness
    top_k=50,  # Top-k sampling
    top_p=0.95,  # Nucleus sampling
    do_sample=True  # Enable sampling
)

# Decode the output tokens to text
output_text = tokenizer.decode(output_tokens[0], skip_special_tokens=True)

# Extract the generated ICD codes (post-processing might be needed)
print("Generated ICD-10 Codes:")
print(output_text)
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

class HuggingFaceClient:
    def __init__(self, model_id):
        # Load the model and tokenizer from Hugging Face
        self.tokenizer = AutoTokenizer.from_pretrained(model_id)
        self.model = AutoModelForCausalLM.from_pretrained(
            model_id,
            torch_dtype=torch.float16,  # Use mixed precision for faster inference
            device_map="auto"          # Automatically allocate model to GPU/CPU
        )

    def chat_completions_create(self, model, messages, max_tokens):
        """
        Simulates the behavior of client.chat.completions.create by generating a response
        using the Hugging Face model.
        """
        # Extract the user prompt from the messages
        prompt = messages[0]["content"]

        # Tokenize the input prompt
        input_ids = self.tokenizer(prompt, return_tensors="pt", truncation=True, max_length=2048).input_ids.to(self.model.device)

        # Generate a response
        outputs = self.model.generate(
            input_ids,
            max_new_tokens=max_tokens,  # Limit the number of new tokens
            temperature=0.7,            # Sampling temperature for diversity
            top_p=0.9,                  # Top-p nucleus sampling for better quality
            do_sample=True              # Enable sampling
        )

        # Decode the generated response
        generated_text = self.tokenizer.decode(outputs[0], skip_special_tokens=True)
        
        # Return the response in a similar structure to OpenAI API
        return {"choices": [{"message": {"content": generated_text}}]}

# Initialize the client
model_id = "meta-llama/Llama-3.2-11B-Vision-Instruct"
client = HuggingFaceClient(model_id)
```
